```python
# 載入所需函示庫
import pandas as pd
import numpy as np
import requests
import googlemaps
from matplotlib import pyplot as plt
from haversine import haversine, Unit
%matplotlib inline
# 設定圖形大小; DPI越大圖越大
plt.rcParams["figure.dpi"] = 60

def convert_to_int(x):
    ans = ''
    for i in x:
        if i.isdigit() or i == '.':
            ans += i
    return(float(ans))
```
